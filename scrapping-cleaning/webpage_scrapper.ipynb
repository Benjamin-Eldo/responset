```python
df_url = pd.read_csv('data/top-1m.csv', header=None, names=['rank', 'url'])
df_url = df_url.drop('rank', axis=1)
```

```python
def get_website_visual(url,filepath, mobile=False):
    try :
        chrome_options = Options()
        chrome_options.add_argument("--headless=new")
        chrome_options.add_argument("--mute-audio")
        chrome_options.add_argument('--hide-scrollbars')
        chrome_options.add_argument('--start-maximized')
        chrome_options.add_argument('--start-fullscreen')
        # chrome_options.add_argument("--window-size=1024,768")
        chrome_options.add_extension('extension/no-cookie.crx')

        if mobile:
            chrome_options.add_argument('--user-agent=Mozilla/5.0 (iPhone; CPU iPhone OS 17_7 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) CriOS/131.0.6778.134 Mobile/15E148 Safari/604.1')
            chrome_options.add_argument("--window-size=393,873")

        driver = webdriver.Chrome(options=chrome_options)
        driver.get("http://"+url)
        driver.save_screenshot(filepath)

        print("Screenshot saved as: "+filepath)
    except Exception as e:
        print("error")
    finally:
        driver.quit()

get_website_visual(df_url['url'][6],"output/output.png", mobile=True)
```

```python
for i in range (762,1010,1):
    filepath_desktop = 'output/desktop/'+df_url['url'][i].replace(".", "_")+'.png'
    filepath_mobile = 'output/mobile/'+df_url['url'][i].replace(".", "_")+'.png'
    get_website_visual(df_url['url'][i], filepath_desktop)
    get_website_visual(df_url['url'][i], filepath_mobile, mobile=True)
    print(i)
```

# Scraping the HTML and CSS code

```python
def get_website_html(url, filepath_html, filepath_css, mobile=False):
    try:
        chrome_options = Options()
        chrome_options.add_argument("--headless=new")
        chrome_options.add_extension('extension/no-cookie.crx')

        if mobile:
            chrome_options.add_argument('--user-agent=Mozilla/5.0 (iPhone; CPU iPhone OS 17_7 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) CriOS/131.0.6778.134 Mobile/15E148 Safari/604.1')
            chrome_options.add_argument("--window-size=393,873")
        
        driver = webdriver.Chrome(options=chrome_options)
        driver.get("http://"+url)

        fileToWrite = open(filepath_html, "w")
        fileToWrite.write(driver.page_source)
        fileToWrite.close()

        styles = driver.execute_script("""
            let css = '';
            for (let sheet of document.styleSheets) {
                try {
                    for (let rule of sheet.cssRules) {
                        css += rule.cssText + '\\n';
                    }
                } catch (e) {
                    console.log('Cannot access stylesheet:', sheet.href);
                }
            }
            return css;
        """)

        with open(filepath_css, "w", encoding="utf-8") as file:
            file.write(styles)

    except Exception as e:
        print("error")
    finally:
        driver.quit()

get_website_html(df_url['url'][2],"output/output.html", "output/output.css")
```

```python
for i in range (1010):
    filepath_desktop_html = 'output/html/'+df_url['url'][i].replace(".", "_")+'_desktop'+'.html'
    filepath_mobile_html = 'output/html/'+df_url['url'][i].replace(".", "_")+'_mobile'+'.html'

    filepath_desktop_css = 'output/css/'+df_url['url'][i].replace(".", "_")+'_desktop'+'.css'
    filepath_mobile_css = 'output/css/'+df_url['url'][i].replace(".", "_")+'_mobile'+'.css'
    get_website_html(df_url['url'][i], filepath_desktop_html, filepath_desktop_css)
    get_website_html(df_url['url'][i], filepath_mobile_html, filepath_mobile_css, mobile=True)
    print(i)
```
