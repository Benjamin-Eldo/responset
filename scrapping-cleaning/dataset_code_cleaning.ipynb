```python
import os
import re
from bs4 import BeautifulSoup


html_names = os.listdir("dataset/code/desktop/html")
```

```python
def remove_javascript(html_code):
    """
    Remove all <script> tags from the HTML code.
    """
    soup = BeautifulSoup(html_code, 'html.parser')
    
    for script in soup.find_all('script'):
        script.decompose()
    
    return str(soup)
```

```python
def remove_link_from_code(content):
    # find every link in the content
    links = re.findall(r"<a\s+href=\"(.*?)\">", content)
    # remove all the links from the content
    for link in links:
        content = content.replace(link, "")

    links = re.findall(r"<link\s+href=\"(.*?)\">", content)
    # remove all the links from the content
    for link in links:
        content = content.replace(link, "")

    #remove all the script tags
    content = re.sub(r"<script>(.*?)</script>", "", content)

    # remove everything in "content" attribute
    content = re.sub(r"content=\"(.*?)\"", "", content)

    soup = BeautifulSoup(content, 'html.parser')
    
    for script in soup.find_all('script'):
        script.decompose()

    for tag in soup.find_all(True): 
        if tag.has_attr('href'):
            del tag['href']
        if tag.has_attr('src'):
            del tag['src']
    
    return str(soup)
```

```python
def remove_non_latin(text):
    """
    Remove every cyrillic and chinese characters from the text.
    """

    chinese_cyrillic_pattern = r'[\u0400-\u04FF\u4E00-\u9FFF]'
    japanese_korean_pattern = (
        r'[\u3040-\u309F\u30A0-\u30FF\u4E00-\u9FFF\uAC00-\uD7AF]'
    )
    
    cleaned_text = re.sub(chinese_cyrillic_pattern, '', text)
    cleaned_text = re.sub(japanese_korean_pattern, '', cleaned_text)
    
    return cleaned_text
```

```python
for html in html_names: 
    with open("dataset/code/desktop/html/" + html, "r") as f:
        print(html)
        content = f.read()
        content = remove_link_from_code(content)
        content = remove_non_latin(content)
        with open("dataset/code/desktop/html_clean/" + html, "w") as f:
            f.write(content)
```

```python
with open("dataset/code/desktop/html/" + html_names[0], "r") as f:
    content = f.read()
    content = remove_javascript(content)
    content = remove_link_from_code(content)
    print(remove_non_latin(content))
```
