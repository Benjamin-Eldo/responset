```python
import os
import re
from bs4 import BeautifulSoup


html_names = os.listdir("dataset/code/desktop/html")
```

```python
def remove_javascript(html_code):
    """
    Supprime toutes les balises <script> et leur contenu dans un code HTML donné.

    :param html_code: str, le code HTML à nettoyer
    :return: str, le code HTML sans les balises <script>
    """
    # Parser le code HTML avec BeautifulSoup
    soup = BeautifulSoup(html_code, 'html.parser')
    
    # Trouver et supprimer toutes les balises <script>
    for script in soup.find_all('script'):
        script.decompose()
    
    # Retourner le code HTML nettoyé
    return str(soup)
```

```python
def remove_link_from_code(content):
    # find every link in the content
    links = re.findall(r"<a\s+href=\"(.*?)\">", content)
    # remove all the links from the content
    for link in links:
        content = content.replace(link, "")

    links = re.findall(r"<link\s+href=\"(.*?)\">", content)
    # remove all the links from the content
    for link in links:
        content = content.replace(link, "")

    #remove all the script tags
    content = re.sub(r"<script>(.*?)</script>", "", content)

    # remove everything in "content" attribute
    content = re.sub(r"content=\"(.*?)\"", "", content)

    # Parser le code HTML avec BeautifulSoup
    soup = BeautifulSoup(content, 'html.parser')
    
    # Trouver et supprimer toutes les balises <script>
    for script in soup.find_all('script'):
        script.decompose()

    for tag in soup.find_all(True):  # Parcourt toutes les balises
        if tag.has_attr('href'):
            del tag['href']  # Supprime les liens dans href
        if tag.has_attr('src'):
            del tag['src']  # Supprime les liens dans src
    
    # Retourner le code HTML nettoyé
    return str(soup)
```

```python
for html in html_names: 
    with open("dataset/code/desktop/html/" + html, "r") as f:
        print(html)
        content = f.read()
        # remove all URL and webpage link from the content using regex
        content = remove_link_from_code(content)
        # save the content to a new file
        with open("dataset/code/desktop/html_clean/" + html, "w") as f:
            f.write(content)

        

```
